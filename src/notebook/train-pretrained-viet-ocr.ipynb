```python
import json
import os
import matplotlib.pyplot as plt
from vietocr.tool.predictor import Predictor
from vietocr.tool.config import Cfg
from vietocr.tool.translate import build_model
from vietocr.model.trainer import Trainer

# Preprocess labels
json_path = '/kaggle/input/dataset-ocr/dataset/labels.json'
data_root = '/kaggle/input/dataset-ocr/dataset/data'
output_train_txt = '/kaggle/working/train_labels.txt'
output_valid_txt = '/kaggle/working/valid_labels.txt'

# Load JSON data (flat structure: {"filename": "text"})
with open(json_path, 'r') as f:
    labels = json.load(f)

items = list(labels.items())
split_idx = int(0.8 * len(items))  # 80-20 train-validation split
train_items = items[:split_idx]
valid_items = items[split_idx:]

# Write train labels
with open(output_train_txt, 'w') as f:
    for img_name, text in train_items:
        img_path = os.path.join(data_root, img_name)
        f.write(f"{img_path}\t{text}\n")

# Write validation labels
with open(output_valid_txt, 'w') as f:
    for img_name, text in valid_items:
        img_path = os.path.join(data_root, img_name)
        f.write(f"{img_path}\t{text}\n")

# Create merged configuration file
config_content = """
# Vocabulary (Vietnamese characters from base.yml)
vocab: 'aAàÀảẢãÃáÁạẠăĂằẰẳẲẵẴắẮặẶâÂầẦẩẨẫẪấẤậẬbBcCdDđĐeEèÈẻẺẽẼéÉẹẸêÊềỀểỂễỄếẾệỆfFgGhHiIìÌỉỈĩĨíÍịỊjJkKlLmMnNoOòÒỏỎõÕóÓọỌôÔồỒổỔỗỖốỐộỘơƠờỜởỞỡỠớỚợỢpPqQrRsStTuUùÙủỦũŨúÚụỤưƯừỪửỬữỮứỨựỰvVwWxXyYỳỲỷỶỹỸýÝỵỴzZ0123456789!"#$%&''()*+,-./:;<=>?@[\]^_`{|}~ '

# Device
device: cuda:0

# Pre-trained weights (from vgg-transformer.yml)
pretrain: https://vocr.vn/data/vietocr/vgg_transformer.pth
weights: https://vocr.vn/data/vietocr/vgg_transformer.pth

# CNN backbone (from vgg-transformer.yml)
backbone: vgg19_bn
cnn:
    pretrained: True
    ss:
        - [2, 2]
        - [2, 2]
        - [2, 1]
        - [2, 1]
        - [1, 1]
    ks:
        - [2, 2]
        - [2, 2]
        - [2, 1]
        - [2, 1]
        - [1, 1]
    hidden: 256

# Transformer settings (from base.yml)
seq_modeling: transformer
transformer:
    d_model: 256
    nhead: 8
    num_encoder_layers: 6
    num_decoder_layers: 6
    dim_feedforward: 2048
    max_seq_length: 1024
    pos_dropout: 0.1
    trans_dropout: 0.1

# Optimizer (from base.yml)
optimizer:
    max_lr: 0.0003
    pct_start: 0.1

# Trainer settings (from base.yml, adjusted for smaller dataset)
trainer:
    batch_size: 32
    print_every: 100
    valid_every: 500
    iters: 2000
    epochs: 50  # Changed from iters to epochs for consistency
    export: ./weights/transformerocr.pth
    checkpoint: ./checkpoint/transformerocr_checkpoint.pth
    log: ./train.log
    metrics: null  # Enable validation metrics on 100 samples

# Dataset settings (updated for Kaggle paths)
dataset:
    name: custom_dataset
    data_root: /kaggle/input/dataset-ocr/dataset/data
    train_annotation: /kaggle/working/train_labels.txt
    valid_annotation: /kaggle/working/valid_labels.txt
    image_height: 32
    image_min_width: 32
    image_max_width: 512

# Data loader (from base.yml)
dataloader:
    num_workers: 3
    pin_memory: True

# Data augmentation (from base.yml)
aug:
    image_aug: true
    masked_language_model: true

# Predictor (from base.yml)
predictor:
    beamsearch: False

# Quiet mode (from base.yml)
quiet: False
"""
with open('/kaggle/working/custom_vgg_transformer.yml', 'w') as f:
    f.write(config_content)

# Load configuration and train
config = Cfg.load_config_from_file('/kaggle/working/custom_vgg_transformer.yml')
trainer = Trainer(config)
trainer.train()

# Evaluate
trainer.visualize_prediction()
trainer.precision()
```

```python
import os
import subprocess
from IPython.display import FileLink, display

def download_file(path, download_file_name):
    os.chdir('/kaggle/working/')
    zip_name = f"/kaggle/working/{download_file_name}.zip"
    command = f"zip {zip_name} {path} -r"
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    if result.returncode != 0:
        print("Unable to run zip command!")
        print(result.stderr)
        return
    display(FileLink(f'{download_file_name}.zip'))
```

```python
import re
import matplotlib.pyplot as plt

# Raw log text
log = """
iter: 000100 - train loss: 1.217 - lr: 1.57e-04 - load time: 2.87 - gpu time: 59.08
iter: 000200 - train loss: 0.914 - lr: 3.00e-04 - load time: 1.64 - gpu time: 59.94
iter: 000300 - train loss: 0.846 - lr: 2.98e-04 - load time: 1.70 - gpu time: 61.46
iter: 000400 - train loss: 0.820 - lr: 2.91e-04 - load time: 1.68 - gpu time: 61.10
iter: 000500 - train loss: 0.801 - lr: 2.80e-04 - load time: 2.76 - gpu time: 60.89
iter: 000600 - train loss: 0.790 - lr: 2.65e-04 - load time: 1.70 - gpu time: 61.12
iter: 000700 - train loss: 0.777 - lr: 2.46e-04 - load time: 1.99 - gpu time: 61.25
iter: 000800 - train loss: 0.773 - lr: 2.25e-04 - load time: 1.67 - gpu time: 61.11
iter: 000900 - train loss: 0.769 - lr: 2.01e-04 - load time: 1.76 - gpu time: 61.13
iter: 001000 - train loss: 0.761 - lr: 1.76e-04 - load time: 2.56 - gpu time: 61.20
iter: 001100 - train loss: 0.756 - lr: 1.50e-04 - load time: 1.77 - gpu time: 60.94
iter: 001200 - train loss: 0.755 - lr: 1.24e-04 - load time: 1.67 - gpu time: 61.08
iter: 001300 - train loss: 0.751 - lr: 9.85e-05 - load time: 1.75 - gpu time: 61.12
iter: 001400 - train loss: 0.749 - lr: 7.48e-05 - load time: 2.66 - gpu time: 61.16
iter: 001500 - train loss: 0.746 - lr: 5.34e-05 - load time: 1.76 - gpu time: 61.22
iter: 001600 - train loss: 0.743 - lr: 3.49e-05 - load time: 2.03 - gpu time: 61.27
iter: 001700 - train loss: 0.744 - lr: 2.00e-05 - load time: 1.80 - gpu time: 61.19
iter: 001800 - train loss: 0.743 - lr: 8.96e-06 - load time: 1.86 - gpu time: 61.23
iter: 001900 - train loss: 0.742 - lr: 2.23e-06 - load time: 2.57 - gpu time: 61.24
iter: 002000 - train loss: 0.742 - lr: 1.43e-09 - load time: 1.80 - gpu time: 61.20
"""

# Parse training iterations and losses
train_iters = []
train_losses = []

for line in log.strip().split('\n'):
    if "train loss" in line:
        match = re.search(r"iter: (\d+) - train loss: ([\d.]+)", line)
        if match:
            train_iters.append(int(match.group(1)))
            train_losses.append(float(match.group(2)))

# Plot
plt.figure(figsize=(10, 5))
plt.plot(train_iters, train_losses, marker='o', color='blue', label="Train Loss")
plt.title("Training Loss over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Train Loss")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
